**Perform Multi-varient Analysis to understand co-relation**

```python
company_list
```

```python
apple = pd.read_csv(company_list[0])
amazon = pd.read_csv(company_list[1])
google = pd.read_csv(company_list[2])
msft = pd.read_csv(company_list[3])
```

```python
closing_data = pd.DataFrame()
```

```python
closing_data['Apple'] = apple['close']
closing_data['Amazon'] = amazon['close']
closing_data['Google'] = google['close']
closing_data['Msft'] = msft['close']
```

```python
closing_data.head()
```

```python
sns.pairplot(closing_data)
```

```python
closing_data.corr()
```

```python
sns.heatmap(closing_data.corr(), annot=True)
```

**Perform co-relation Analysis**

```python
closing_data['Apple'].shift(1)
```

# (x-y)/y * 100

# shift(1) is used for calculating x-y

```python
(closing_data['Apple'] - closing_data['Apple'].shift(1) ) / closing_data['Apple'].shift(1) * 100
```

```python
for cd in closing_data:
  closing_data[cd + ' pct_change']= (closing_data[cd] - closing_data[cd].shift(1) ) / closing_data[cd].shift(1) * 100
```

```python
closing_data.columns
```

```python
closing_pct = closing_data[['Apple pct_change',
       'Amazon pct_change', 'Google pct_change', 'Msft pct_change']]
```

```python
closing_pct
```

```python
g = sns.PairGrid(data = closing_pct)
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```

```python
closing_pct.corr()
```

```python
sns.heatmap(closing_pct.corr(), annot=True)
```
